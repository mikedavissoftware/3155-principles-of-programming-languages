Terminals: { Z, Succ(.) }

Start symbol: N

N => Z
N => Succ(N) 

```scala
sealed trait N // declare a nonterminal
case object Z extends N // N => Z
case class Succ(n: N) extends N // N => Succ(N)
```

```scala
val one = Succ(Z)
val two = Succ(Succ(Z))
val three = Succ(Succ(Succ(Z)))
one == two
Succ(two) == three
```
