# Dictionary of Big Step Semantics

When developing our own programming language, it is useful to start with inductive definitions of nonterminals and terminals.

Nonterminals are the base domains, from which all other types are derived (extended). For the purposes of our language, nonterminals include:
- $ \mathbb{R} $ for real numbers
- $ \mathbb{B} $ for booleans
- $ \mathbb{C} $ for closures (these represent functions)
- $ \mathbb{P} $ for references (the P stands for "pointers")
- $ \textbf{error} $ for errors

To adequately lay out the process of evaluating various expression types, we need __big step semantics__. Through doing this, we can easily visualize the steps an interpreter needs to take in order to actually execute the evaluation.

The basic layout of how we notate our __inference rules__ is as follows:

$$\begin{array}{c}
\text{premises that must hold} \\
\hline
\text{conclusions that can be drawn} \\
\end{array}$$

Before we write the rules for all out, let's go over some __symbology__:

$ \sigma $ = environment <br>
$ \text{e} $ = some expression ($ \text{e1} $, $ \text{e2} $) <br>
$ \text{s} $ = some string <br>
$ \text{c} $ = some contant (real number) <br>
$ \in $ = "is contained within"

<hr>

### Constants

$$ \begin{array}{c}
\\
\hline
\textbf{eval}\left(\texttt{Const(f)}, \sigma\right) = f \\
\end{array}\ \text{(Const)} $$

<hr>

### Identifiers

$$\begin{array}{c}
s \in \text{Domain}(\sigma)\\
\hline
\textbf{eval}(\texttt{Ident(s)},\sigma) = \sigma(s) \\
\end{array}\ (\text{Ident})$$

$$ \begin{array}{c}
 s \not\in \text{Domain}(\sigma) \\
 \hline
 \textbf{eval}(\texttt{Ident(s)}, \sigma) = \mathbf{error} \\
 \end{array} (\text{Ident-ERROR}) $$

<hr>

### Addition

$$\begin{array}{c}
\textbf{eval}(\texttt{e1}, \sigma) = c_1,\ \textbf{eval}(\texttt{e2}, \sigma) =  c_2\\
\hline
\textbf{eval}\left( \texttt{Plus(e1,e2)}, \sigma\right)  =  (c_1 + c_2) \\
\end{array} (\text{Plus}) $$

<hr>

### Subtraction

$$\begin{array}{c}
\textbf{eval}(\texttt{e1}, \sigma) = c_1,\ \textbf{eval}(\texttt{e2}, \sigma) =  c_2 \\
\hline
\textbf{eval}\left(\texttt{Minus(e1, e2)}, \sigma\right) = (c_1 - c_2 ) \\
\end{array} (\text{Minus}) $$

<hr>

### Multiplication

$$\begin{array}{c}
\textbf{eval}(\texttt{e1}, \sigma) = c_1,\ \textbf{eval}(\texttt{e2}, \sigma) =  c_2 \\
\hline
\textbf{eval}\left(\texttt{Mult(e1, e2)}, \sigma\right) = (c_1 \times c_2) \\
\end{array} (\text{Mult}) $$

<hr>

### Division

$$\begin{array}{c}
\textbf{eval}(\texttt{e1}, \sigma) = c_1,\ \textbf{eval}(\texttt{e2}, \sigma) =  c_2,\ \color{red}{c_2 \not= 0}\\
\hline
\textbf{eval}\left( \texttt{Div(e1, e2)}, \sigma \right) = (\frac{c_1}{c_2} ) \\
\end{array} (\text{Div}) $$

$$\begin{array}{c}
\textbf{eval}(\texttt{e1}, \sigma) =  c_1,\ \textbf{eval}(\texttt{e2}, \sigma) = c_2,\ {c_1 \in \mathbb{R}, c_2 \in \mathbb{R}},\ c_2 = 0\\
\hline
\textbf{eval}\left(\texttt{Div(e1, e2)}, \sigma\right) = \mathbf{error} \\
\end{array} (\text{Div-ERROR}) $$

<hr>

### Logarithms

$$\begin{array}{c}
\textbf{eval}\left( \texttt{e}, \sigma\right) = c,\ \color{red}{c > 0} \\
\hline 
\textbf{eval}(\texttt{Log(e)}, \sigma) = \log(c) \\
\end{array}\ \text{(Log)}$$

$$\begin{array}{c}
\textbf{eval}(\texttt{e}, \sigma) = c,\ {c \in \mathbb{R}},\ {c \leq 0} \\
\hline 
\textbf{eval}(\texttt{Log(e)}, \sigma) =  \mathbf{error} \\
\end{array}\ \text{(Log-ERROR)}$$
