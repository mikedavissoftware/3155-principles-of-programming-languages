# A Dictionary for Abstract Inductive Grammar

## Simple Arithmetic Grammar
#### Below is the grammar for arithmetic grammar, all of which stems from expressions:

$$\begin{array}{rcc}
\textbf{Expr} & \rightarrow & Const(\textbf{Double}) \\
& | & Ident(\textbf{Identifier}) \\
& | & Plus( \textbf{Expr}, \textbf{Expr}) \\
& | & Minus( \textbf{Expr}, \textbf{Expr}) \\
& | & Mult(\textbf{Expr}, \textbf{Expr}) \\
& | & Div(\textbf{Expr}, \textbf{Expr}) \\
& | & Log(\textbf{Expr}) \\
& | & Exp(\textbf{Expr}) \\
& | & Sine(\textbf{Expr}) \\
& | & Cosine(\textbf{Expr}) \\\\
\end{array}$$

```scala
// Nonterminals
sealed trait Expr // Short for Expression

// Terminals
case class Const(d: Double) // Double is a class built directly into Scala
case class Ident(s: String) // String is a class built directly into Scala
case class Plus(e1: Expr, e2: Expr)
case class Minus(e1: Expr, e2: Expr)
case class Mult(e1: Expr, e2: Expr)
case class Div(e1: Expr, e2: Expr)
case class Log(e: Expr)
case class Exp(e: Expr)
case class Sine(e: Expr)
case class Cosine(e: Expr)
```

## More Complex Grammar
#### Below is a complete grammar for a simple while programming language:

Note: * is a Kleene star, which quantifies zero or more occurences of the previous expression

$$\begin{array}{rcll}
\textbf{Program} & \rightarrow & \textbf{Declaration}^*\ \textbf{Statement}^*\ ReturnStmt(\textbf{Expr}) \\[5pt]
\textbf{Declaration} & \rightarrow & VarDecl( \textbf{Identifier}, \textbf{Expr} )\\[5pt]
\textbf{Statement} & \rightarrow & Assign( \textbf{Identifier}, \textbf{Expr} ) & \text{assign identifier to expression} \\
& | & While(\textbf{CondExpr}, \textbf{Statement}^*) & \text{the condition  and the list of statements: note the Kleene star} \\
& | & IfThenElse(\textbf{CondExpr}, \textbf{Statement}^*, \textbf{Statement}^*) & \text{can you interpret this rule as an exercise?}\\
& | & ReturnStmt(\textbf{Expr}) & \text{return an expression: we already saw this.} \\[5pt]
\textbf{Expr} & \rightarrow & Const(\textbf{Double}) \\
& |  & Ident(\textbf{Identifier}) \\
& | & Plus( \textbf{Expr}, \textbf{Expr}^+) & \textbf{Expr}^+ \text{denotes one or more occurrences of an expression}\\
& | & Minus( \textbf{Expr}, \textbf{Expr}^+) & e_1 - e_2 - e_3 - e_4 \cdots \\
& | & Mult(\textbf{Expr}, \textbf{Expr}^+) & e_1 * e_2 * e_3 * \cdots \\
& | & Div(\textbf{Expr}, \textbf{Expr}) \\
& | & Log(\textbf{Expr}) \\
& | & Exp(\textbf{Expr}) \\
& | & Sine(\textbf{Expr}) \\
& | & Cosine(\textbf{Expr}) \\[5pt]
\textbf{CondExpr} & \rightarrow & ConstTrue & \text{constant boolean true}\\
& | & ConstFalse & \text{constant boolean false} \\
& | & Geq( \textbf{Expr} ,\textbf{Expr}) & \ e_1 \geq e_2 \\
& | & Leq (\textbf{Expr}, \textbf{Expr}) & \ e_1 \leq e_2 \\
& | & Eq(\textbf{Expr} , \textbf{Expr}) &  e_1 == e_2 \\
& | & And(\textbf{CondExpr}, \textbf{CondExpr}) & \ c_1\ \text{and}\ c_2 \\
& | & Or(\textbf{CondExpr}, \textbf{CondExpr}) & \ c_1\ \text{or}\ c_2 \\
& | & Not(\textbf{CondExpr}) &  \ \text{not}\ c_1 \\[5pt]
\textbf{Double} & \rightarrow & \cdots\ |\  -2\ |\ -1\ |\ 0\ |\ 1\ |\ 2\ |\ \cdots \\
\textbf{Identifier} & \rightarrow & [a-z\ A-Z][a-z\ A-Z\ 0-9\ \_]* & \text{this is a string}
\end{array}$$

```scala
// In proper order...

sealed trait Expr
case class Const(d: Double) extends Expr 
// 1. We cheated a bit and allowed all floating point numbers
// Also, this deviates from the grammar
case class Ident(s: String) extends Expr
// 2. We allow any string to be an identifier for now instead of the regular expression shown in the grammar.
case class Plus(e1: Expr, e2: Expr ) extends Expr
case class Minus(e1: Expr, e2: Expr) extends Expr
case class Mult(e1: Expr, e2: Expr) extends Expr
case class Div(e1: Expr, e2: Expr) extends Expr
case class Negate(e: Expr) extends Expr
case class Log(e: Expr) extends Expr
case class Exp(e: Expr) extends Expr
case class Sine(e: Expr) extends Expr
case class Cosine(e: Expr) extends Expr

sealed trait CondExpr
case object ConstTrue extends CondExpr
case object ConstFalse extends CondExpr
case class Geq(e1: Expr, e2: Expr) extends CondExpr
case class Leq(e1: Expr, e2: Expr) extends CondExpr
case class Eq(e1: Expr, e2: Expr) extends CondExpr
case class And(c1: CondExpr, c2: CondExpr) extends CondExpr
case class Or(c1: CondExpr, c2: CondExpr) extends CondExpr
case class Not(c: CondExpr) extends CondExpr

sealed trait Declaration
sealed trait Statement
sealed trait MythonProgram

case class Program(decls: List[Declaration], stmts: List[Statement], returnAtEnd: Expr) extends MythonProgram // We stripped the ReturnStmt tag since it is redundant
case class VarDecl(identifier: String, rhsExpr: Expr) extends Declaration
case class AssignStmt(identifier: String, rhsExpr: Expr) extends Statement
case class WhileStmt(cond: CondExpr, stmts: List[Statement]) extends Statement
case class IfThenElseStmt(cond: CondExpr, stmtsThen: List[Statement], stmtsElse: List[Statement]) extends Statement
case class ReturnStmt(retExpr: Expr) extends Statement

sealed trait Case

case class Switch(e: Expr, cases: List[Case]) extends Statement
case class SwitchCase(e: Expr, stmts: List[Statement]) extends Case
case class DefaultCase(stmts: List[Statement]) extends Case
```
